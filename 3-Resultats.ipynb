```python
import Levenshtein
import random
```

### Files

```python
database_file = "data/rockyou.txt"
generated_file = "data/generated_pwd.txt"
```

```python
database =  set()
with open(database_file, errors='ignore') as f:
    for line in f.readlines():
        database.add(line.replace("\n", ""))
database = list(database)
random.shuffle(database)


all_generated = set()
with open(generated_file) as f:
    for line in f.readlines():
        all_generated.add(line.replace("\n", ""))
        
all_generated = list(all_generated)

print("Taille de la base de données: ", len(database))
print("Taille des mots de passes générés: ", len(all_generated))
```

```python
part_database = database[:int(0.01*len(database))]
part_generated = all_generated[:int(0.16*len(all_generated))]
print("part_database", len(part_database))
print("part_generated", len(part_generated))
cracked = list(set(part_database).intersection(part_generated))

print(len(set(cracked)))
print(len(set(cracked))/len(part_database))
```

```python
print(len(set(database).intersection(all_generated)))
```

### Résultats

```python
# seuil d'acceptation (0 caractères = match parfait, 1 caractère près ...)
seuil = 1
cracked = []

i = 0

for mdp in database:
    for generated in all_generated:
        dist = Levenshtein.distance(generated.strip(), mdp.strip())
        if dist <= seuil:
            cracked.append(generated)
            print(generated)
    i += 1
    
    print(i)
    
print(len(cracked))
```
